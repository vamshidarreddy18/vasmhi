###observations from correlation plots and _coeffcients -Between x and y all the x variables are showing to high correlation strengths,highest being between hp and mpg -Therefore this dataset qualifies for building a multiple linear regression model to predict mpg

the high correlation among x columns is not desirable as it might lead to multicollinearity problem

```python
model = smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
```

```python
model.summary()
```

###observations from model summary
- the r squared and adjusted r squared values are good about 75% of variablility in y is explained by x columns
- the probability value with respect to f statistics is close to zero indicating that all or someo of x columns are significant

```python
cars=pd.DataFrame(cars,columns=['HP','VOL','SP','WT','MPG']) 
cars.head()
```

```python
cars.head()
```

```python
df1 = pd.DataFrame()
df1["actual_y1"] = cars["MPG"]
df1.head()
```

```python
pred_y1 = model.predict(cars.iloc[:,0:4])
df1["pred_y1"] = pred_y1
df1.head()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print("MSE :", mean_squared_error(df1["actual_y1"],df1["pred_y1"]))
```
